```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv("new_csv.csv")
df.head()
```

```python
df2=df.copy()
df2.head()
```

```python
df2["region_shape_attributes"] = df2["region_shape_attributes"].apply(lambda x : dict(eval(x)) )
df3 = df2["region_shape_attributes"].apply(pd.Series )
df3
```

```python
df.shape
```

```python
result = pd.concat([df, df3], axis=1)
result
```

```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from sklearn.cluster import KMeans
%matplotlib inline
```

```python
plt.scatter(result.width, result.height)
```

```python
final_data = result[['width', 'height']]
final_data
```

```python
final_data.info()
```

```python
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', max_iter=300, n_init=10, random_state=0)
    kmeans.fit(final_data)
    wcss.append(kmeans.inertia_)
plt.plot(range(1, 11), wcss)
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

### Desired Number of Clusters - 4

```python
kmeans = KMeans(n_clusters=4, init='k-means++', max_iter=300, n_init=10, random_state=0)
pred_y = kmeans.fit_predict(final_data)
plt.scatter(final_data.width, final_data.height)
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=300, c='red')
plt.show()
```
